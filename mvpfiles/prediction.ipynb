```python
import pandas as pd

totalData = None

for i in range(1987, 2013):
    data = pd.read_csv("./trainData/"+ str(i+1) + "mvpVotes.csv")
    if(i == 1987):
        totalData = pd.DataFrame(columns = data.columns)
    totalData = pd.concat([totalData, data], ignore_index=True)

for column in data.columns:
    totalData.rename(columns={column:data.loc[0].at[column]}, inplace=True)

totalData = totalData.drop(totalData.index[0])
totalData = totalData.drop(totalData.columns[[0]], axis=1)
totalData['3P%'] = totalData['3P%'].fillna(0)
totalData = totalData[totalData['Age'] != 'Age']



display(totalData)
#totalData.info()
```

```python
from sklearn.model_selection import train_test_split

names = totalData['Player']
#display(names)
x = totalData.drop(columns=['Player', 'Share', 'Rank', 'Tm', 'First', 'Pts Won', 'Pts Max'])
#display(x)
y = totalData['Share']

xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size=0.2, random_state=100)
```

```python
from sklearn.ensemble import RandomForestRegressor

randomForest = RandomForestRegressor(n_estimators=100, random_state=2)
randomForest.fit(xTrain, yTrain)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

yPred = randomForest.predict(xTest)
err = mean_squared_error(yTest, yPred)
r2err = r2_score(yTest, yPred)

print(f"Mean Squared Error: {err:.4f}")
print(f"R^2 Score: {r2err:.4f}")

#display(yPred)

results = yTest.to_frame(name = 'yTest')
results['yPred'] = yPred
results['PlayerName'] = [names.loc[x] for x in results.index.values]
display(results)


```
