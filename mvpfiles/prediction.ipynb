```python
import pandas as pd

totalData = None

for i in range(1987, 2013):
    data = pd.read_csv("./trainData/"+ str(i+1) + "mvpVotes.csv")
    if(i == 1987):
        totalData = pd.DataFrame(columns = data.columns)
    totalData = pd.concat([totalData, data], ignore_index=True)

for column in data.columns:
    totalData.rename(columns={column:data.loc[0].at[column]}, inplace=True)

totalData = totalData.drop(totalData.index[0])
totalData = totalData.drop(totalData.columns[[0]], axis=1)
totalData['3P%'] = totalData['3P%'].fillna(0)
display(totalData)
totalData.info()





```
